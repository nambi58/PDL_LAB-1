## PART-1 Write a method for sigmoid functions

```python
import numpy as np 
import math
```

```python
def mysigmoid(x):
    y= 1/(1+(math.exp(-x)))
    return y
```

```python
mysigmoid(4)
```

```python
x=[1,2,3]
for i in x:
    print(mysigmoid(i))
```

```python
def mysigmoidi(x):
    y= 1/(1+(np.exp(-x)))
    return y
```

```python
mysigmoidi(4)
```

```python
x=[1,2,3]
for i in x:
    Z=(mysigmoidi(i))
    print(Z)
```

## PART-2 Gradient or derivative of sigmoid function

```python
def sig_derivaties(s):
    s_derivate = s*(1-s)
    return s_derivate
```

```python
sig_derivaties(mysigmoid(4))
```

### PART-4 Write a method normalizeRows()

```python
def normalizeRows(x):
    x_norm = np.linalg.norm(x, ord= 2, axis= 1, keepdims = True)
    
    x= x/x_norm
    
    return x
```

```python
x= np.array([[0,3,4],[1,6,4]])
print("normalizeRows(x) = "+ str(normalizeRows(x)))
```

### PART-6 Implement L1 and L2 loss functions

```python
def L1(yhat, y):
 loss = np.sum(abs(y - yhat), axis = 0, keepdims = True)
 return loss
y = np.array([1, 0, 0, 1, 1])
yhat = np.array([.9, 0.2, 0.1, .4, .9])
print("L1 = " + str(L1(yhat,y)))
```

```python
def L2(yhat, y):
 loss = np.dot((y - yhat),(y - yhat))
 return loss
y = np.array([1, 0, 0, 1, 1])
yhat = np.array([.9, 0.2, 0.1, .4, .9])
print("L2 = " + str(L2(yhat,y)))
```
